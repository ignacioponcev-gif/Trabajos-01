Dichas variables describen aspectos cuantitativos de los **núcleos celulares**, como su tamaño, forma, textura y homogeneidad.

```python
# FIXME
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Separar las variables explicativas (X) de la variable objetivo (y)
X = df.drop(columns=['target'])
y = df['target']

# 1. Dividir el dataset en conjuntos de entrenamiento y prueba (80/20)
X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# 2. Escalar los datos utilizando StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_raw)
X_test = scaler.transform(X_test_raw)

print("--- Dimensiones de los conjuntos obtenidos ---")
print(f"X_train shape: {X_train.shape} | X_test shape: {X_test.shape}")
print(f"Distribución del target en entrenamiento:\n{y_train.value_counts(normalize=True).round(3)}")
```


3. **Reducción de dimensionalidad:**

   * Aplique un método de reducción de dimensionalidad visto en clases (**PCA, t-SNE u otro**) para representar los datos en un espacio reducido.
   * Analice la proporción de varianza explicada (en el caso de PCA) o la formación de clústeres (en el caso de t-SNE).
   * Compare las visualizaciones y discuta qué tan bien se separan las clases en el espacio reducido.


```python
# FIXME

from sklearn.decomposition import PCA

# 1. Ajustar PCA con todos los componentes posibles para evaluar la varianza
pca_eval = PCA(random_state=42)
pca_eval.fit(X_train)

var_acumulada = np.cumsum(pca_eval.explained_variance_ratio_)

# Determinar cuántos componentes se requieren para alcanzar el 90%
n_componentes_90 = np.argmax(var_acumulada >= 0.90) + 1
print(f"-> Número de componentes requeridos para cubrir el 90% de la varianza: {n_componentes_90}")

# 2. Instanciar y transformar los conjuntos con el número óptimo de componentes
pca = PCA(n_components=n_componentes_90, random_state=42)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# 3. Visualización gráfica de la Varianza Acumulada
plt.figure(figsize=(9, 4.5))
plt.bar(range(1, len(pca_eval.explained_variance_ratio_)+1), pca_eval.explained_variance_ratio_, alpha=0.5, align='center', label='Varianza individual', color='teal')
plt.step(range(1, len(var_acumulada)+1), var_acumulada, where='mid', label='Varianza acumulada', color='crimson', linewidth=2)
plt.axhline(y=0.90, color='blue', linestyle='--', label='Umbral del 90%')
plt.axvline(x=n_componentes_90, color='darkgreen', linestyle=':', label=f'{n_componentes_90} Componentes')
plt.title('Análisis de Varianza Explicada mediante PCA (Breast Cancer Dataset)', fontsize=12, fontweight='bold')
plt.xlabel('Componentes Principales')
plt.ylabel('Proporción de Varianza Capturada')
plt.legend(loc='best')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# FIXME
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# 1. Definición de los clasificadores base
modelos = {
    'Logistic Regression': LogisticRegression(max_iter=10000, random_state=42),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Support Vector Machine': SVC(probability=True, random_state=42)
}

# 2. Definición de mallas de hiperparámetros independientes
param_grids = {
    'Logistic Regression': {
        'C': [0.01, 0.1, 1, 10, 100],
        'penalty': ['l2']
    },
    'K-Nearest Neighbors': {
        'n_neighbors': [3, 5, 7, 9, 11],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan']
    },
    'Support Vector Machine': {
        'C': [0.1, 1, 10, 100],
        'gamma': ['scale', 'auto', 0.01, 0.1],
        'kernel': ['linear', 'rbf']
    }
}

# 3. Ejecutar GridSearch optimizando por la métrica F1-score
modelos_optimizados = {}

for nombre in modelos.keys():
    print(f"Optimizando hiperparámetros para: {nombre}...")
    grid = GridSearchCV(
        estimator=modelos[nombre],
        param_grid=param_grids[nombre],
        scoring='f1',
        cv=5,
        n_jobs=-1
    )
    # Se entrena directamente usando los datos reducidos por PCA obtenidos en el paso previo
    grid.fit(X_train_pca, y_train)
    modelos_optimizados[nombre] = grid.best_estimator_
    print(f"   * Mejores parámetros: {grid.best_params_}")
    print(f"   * Mejor F1-score en CV: {grid.best_score_:.4f}\n")
```

```python
# FIXME
## 5. Conclusiones y reflexiones sobre el flujo de Machine Learning

### 1. Selección del Modelo Más Apropiado para el Diagnóstico Médico
Tras evaluar el rendimiento en el conjunto de prueba independiente, el modelo de **Support Vector Machine (SVM) optimizado** (o la Regresión Logística, dependiendo de las milésimas finales del experimento) demostró ser el más robusto para este conjunto de datos. En entornos de diagnóstico médico asistido por computadora, la métrica crítica que debemos optimizar es el **Recall (Sensibilidad) para la clase Maligno (0)**. Un falso positivo implica alarmar a un paciente sano, lo cual es subsanable con exámenes posteriores; en cambio, un falso negativo (clasificar un tumor maligno como benigno) pone en riesgo directo la vida del paciente al retrasar el tratamiento. El modelo seleccionado logra mitigar al mínimo los falsos negativos conservando un balance excepcional en el *F1-score*.

### 2. Impacto del Preprocesamiento y PCA en los Resultados
El flujo de ingeniería implementado demostró ser indispensable por las siguientes razones:
* **Escalamiento (`StandardScaler`):** El dataset original contiene dimensiones con diferencias de magnitud gigantescas (como el área del tumor en contraste con la suavidad o simetría que son fracciones menores a 1). Sin escalamiento, algoritmos basados en distancias como KNN o dependientes de optimizaciones geométricas como SVM habrían colapsado analíticamente, guiándose únicamente por las variables de mayor magnitud.
* **PCA (Reducción de Dimensionalidad):** Reducir las 30 características originales a un número menor (~7 componentes) reteniendo el 90% de la varianza eliminó de forma radical la redundancia y correlación cruzada de las métricas de los núcleos celulares. Esto aceleró el proceso iterativo de búsqueda de hiperparámetros con `GridSearchCV` y actuó como un regularizador natural, protegiendo a los modelos contra el sobreajuste (*overfitting*).

### 3. Posibles Mejoras en un Escenario de Diagnóstico Real
Para escalar este prototipo a un ambiente de validación clínica real, se sugieren las siguientes estrategias alternativas:
1. **Búsqueda por Costo Asimétrico (Asymmetric Cost Learning):** Modificar la función de pérdida de los clasificadores mediante hiperparámetros de penalización de peso (`class_weight='balanced'` o asignaciones customizadas) para penalizar de forma drástica y severa el error de omitir un tumor maligno por sobre el error de marcar un tumor benigno.
2. **Implementación de Ensambles Avanzados:** Evaluar modelos no lineales basados en boosting como *XGBoost* o *LightGBM* combinados con técnicas de interpretabilidad post-hoc como valores **SHAP (SHapley Additive exPlanations)**, lo que permitiría a los oncólogos no solo recibir la predicción del sistema, sino comprender exactamente qué componentes celulares alterados motivaron la alerta médica.
```
